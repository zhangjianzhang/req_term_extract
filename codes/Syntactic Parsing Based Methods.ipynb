```python
# define functions to extract NP nodes and words
def get_np_words(node):
    """
    Get the words contained in the NP node
    """
    return ' '.join([p.replace(' ','').replace(')','') for p in re.findall(r'\s[\w\(\),]*\)', str(node))])
def extract_np(node):
    """
    get NP nodes from the parse tree
    """
    if node.label == 'NP':
        print("NP:",get_np_words(node) ) 
        return node
    for child in node.children:
        extract_np(child)
```

```python
# read sample text
with open("Spotify.txt",'r',encoding="utf-8") as t:
    example = t.read()
```

```python
# extract the noun phrase of each sentence
nlp = stanza.Pipeline(lang='en', processors='tokenize,pos,constituency')
doc = nlp(example)
for sentence in doc.sentences:
    print()
    print("Sentence:", sentence.text)
    extract_np(sentence.constituency)
```
